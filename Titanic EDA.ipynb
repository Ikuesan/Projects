```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#Overall survival rate of the Titanic

survival_rate = df["Survived"].mean()
```

```python
round(survival_rate, 4) #rounding off the survival rate to 4 dp
```

```python
survival_counts = df['Survived'].value_counts()
survival_rate = survival_counts / len(df)
```

```python
survival_rate
```

```python
survival_rate.plot(kind="bar", title="Survival rate on the Titanic")
plt.xlabel("Survived")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
#a higher rate of passengers did not survive on the titanic as seen in the bar plot above
```

```python
#How did survival rates differ between classes?

survival_rate_by_class = df.groupby('Pclass')['Survived'].mean() 
```

```python
survival_rate_by_class.plot(kind="bar", title="Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
#Passengers in the class 1 have a higher survival rate than those in other classes which means that passengers in class 1 were more liable to survive than any other 
```

```python
#What was the distribution of ages on the Titanic?

plt.hist(df["Age"], edgecolor="black") 
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
#from the histogram above, passengers within the ages of 16 - 32 (teenagers and young adults) were the most populated on the Titanic
```

```python
#boxplot of age distribution by survival status
df.boxplot(column='Age', by='Survived')
plt.title('Age Distribution by Survival Status')
```

```python
# Passengers that did not survive had a median age of about 29 years, 25% of them were below the age of 21, 
# 75% of them were below the age of 39, the maximum age was about 66, the minimum about one month old with some outliers of 70 - 73 years old
# Passengers that did survive had a median age of about 27 years, 25% of them were below the age of 19, 
# 75% of them were below the age of about 47, the maximum age was 60 years, the minimum 0 year and some outlier age of above 60 to 80 years
```

```python
#Were women and children more likely to survive?

df["Child"] = df["Age"]<16 #creating a new column "child" where the age was less than 16
```

```python
df
```

```python
#subsetting the df for children only to create a new df

children_df = df[df["Child"]==True]
children_df
```

```python
#subsetting the df for females only with ages equal to or above 16 (i.e women)

female_df = df[(df["Sex"] == "female") & (df["Age"] >= 16)]
female_df
```

```python
#mean survival rate for women only

mean_survival_female = female_df["Survived"].mean()
```

```python
#subsetting the df for males only with ages equal to or above 16 (i.e men)

male_df = df[(df["Sex"] == "male") & (df["Age"] >= 16)]

male_df
```

```python
#mean survival rate for men only

mean_survival_male = male_df["Survived"].mean()
```

```python
#mean survival rate for children only

mean_survival_child = children_df["Survived"].mean()
```

```python
#creating a list of the df for men, women and children

groups = ['Females', 'Males', 'Children']
mean_survival_rates = [
    mean_survival_female,
    mean_survival_male,
    mean_survival_child
]
```

```python
plt.bar(groups, mean_survival_rates, color=['purple', 'blue', 'green'])
plt.title("survival rate by gender and age group")
plt.show()
```

```python
#From the bar chart above, women and children were more likely to survive
```

```python
#What was the impact of fare paid on survival?

plt.hist(df["Fare"], edgecolor="black") 
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
```

```python
df.boxplot(column= "Fare", by="Survived")
plt.title('Fare Distribution by Survival Status')
```

```python
#Passengers that survived, paid a median higher fare than those that did not survive. This could have contributed to their survival.
```

```python
#Did family size affect survival chances?

df["Family_size"] = df["SibSp"] + df["Parch"] + 1 #craeting a new column "Family_size"
```

```python
df
```

```python
#mean survival rate by Family size
survival_rate_by_Family_size = df.groupby("Family_size")["Survived"].mean()
```

```python
survival_rate_by_Family_size.plot(kind="bar", title="Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
#Passengers with smaller family sizes aboard the titanic had a higher survival rate than those with larger family sizes aboard the ship
```

```python
#what is the relationship between embarkation point and survival?

survival_rate_by_embarking_point = df.groupby("Embarked")["Survived"].mean() #grouping the df by "embarked" and calculating the mean survival rate
survival_rate_by_embarking_point
```

```python
survival_rate_by_embarking_point.plot(kind="bar", title="survival_rate_by_embarking_point")
plt.xlabel("embarking_point")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
#Passengers that embarked from port Cherbourg (C) had a higher survival rate than any other port
```

```python
#Is there a correlation between age and fare?

plt.scatter(df["Age"], df["Fare"])
```

```python
#from the scatter plot above, there is little to no correlation between age and fare
```

```python
#correlation coefficient

correlation = df["Age"].corr(df["Fare"])
```
